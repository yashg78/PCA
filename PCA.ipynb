# PCA -
Dimension Reduction

```python
import numpy as np
np.random.seed(123)
```

```python
mea_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
```

```python
sample1 = np.random.multivariate_normal(mea_vec1,cov_mat1,50).T
print(sample1)
```

```python
mea_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
```

```python
sample2 = np.random.multivariate_normal(mea_vec2,cov_mat2,50).T
print(sample2)
```

```python
%pylab inline

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize = (10,5))

ax = fig.add_subplot(111,projection = '3d')
plt.rcParams['legend.fontsize'] = 10

ax.plot(sample1[0,:],sample1[1,:],sample1[2,:],'o',markersize = 10,alpha = 0.5,label = 'Class 1')
ax.plot(sample2[0,:],sample2[1,:],sample2[2,:],'^',markersize = 10,alpha = 0.5,label = 'Class 2')
ax.legend(loc = 'upper right')
plt.show()
```

```python
all_sample = np.concatenate((sample1,sample2), axis =1)
print(all_sample)
```

```python
all_sample.T
```

```python
mean_x = np.mean(all_sample[0,:])
mean_y = np.mean(all_sample[1,:])
mean_z = np.mean(all_sample[2,:])

mean_vec = np.array([[mean_x],[mean_y],[mean_z]])
print(mean_vec)
```

```python
cov_mat = np.cov([all_sample[0,:],all_sample[1,:],all_sample[2,:]])
print(cov_mat)
```

```python
eig_val,eig_vec = np.linalg.eig(cov_mat)
print(eig_val)
print(eig_vec)
```

```python
eig_pair = [(np.abs(eig_val[i]) , eig_vec[:,i]) for i in range(len(eig_val))]
print(eig_pair)
```

```python
eig_pair.sort()
eig_pair.reverse()
print(eig_pair[0][1][newaxis].T)
```

```python
req_matrix = np.hstack((eig_pair[0][1][newaxis].T,eig_pair[1][1][newaxis].T))
print(req_matrix)
```

```python
transformed = req_matrix.T.dot(all_sample)
print(transformed.shape)
```

```python
plt.plot(transformed[0,0:40], transformed[1,0:40],
         'o', markersize=7, color='green', alpha=0.5, label='class1')
plt.plot(transformed[0,40:80], transformed[1,40:80],
         '^', markersize=7, color='red', alpha=0.5, label='class2')
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.legend()
plt.title('Transformed samples with class labels')

plt.show()
```
